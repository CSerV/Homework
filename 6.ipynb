```python
#写函数，返回某个元素/对象在一个list中的位置，如果不在，则返回-1.
n=[45,98,33,6,22,2,9,101,55]
x=(int(input('你想知道什么')))
i=0
for nn in n:
    i=i+1
    if nn==x:
        break
print(x,'是第',i,'个')
```

```python
#写函数，可将两个相同长度的list，间隔插入，生成新的list。例如：给两个list，a=[1,2,3,4], b=[5,6,7,8]。则可以生成：[1,5,2,6,3,7,4,8]。
a=[1,2,3,4]
b=[5,6,7,8]
i=0
number=[]
while i<4:
    number.append(a[i])
    number.append(b[i])
    i=i+1
print(number)
```
