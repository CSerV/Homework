# •练习 2：写函数，共n个随机整数，整数范围在m与k之间，（n,m,k由用户输入）。求1：西格玛log(随机整数)，2：西格玛1/log(随机整数)

```python
import random,math

def my_mul(n):
    i = 0
    total_n = 0
    while i < n:
        number = random.randint(m,k)
        a= math.log(number) 
        i = i + 1
        total_n = total_n + a
    return total_n

def my_sum(n):
    i = 0
    total_n = 0
    while i < n:
        number = random.randint(m,k)
        a= math.log(number) 
        i = i + 1
        b=1/a
        total_n = total_n + b
    return total_n

k=int(input('请输入一个大于0的整数，作为界，回车结束。'))
m=int(input('请输入一个大于0的整数，作为界，回车结束。'))
n=int(input('请输入一个大于0的整数，作为个数，回车结束。'))
    
if k<m:
    k,m=m,k
    print(1,my_mul(n))
    print(2,my_sum(n))
else:
    print(1,my_mul(n))
    print(2,my_sum(n))
        
```

```python
import random

def my_mul(k):
    a = random.randint(1,9)
    i=0
    n=0
    m=0
    while i<k:
        n += a*(10**i)
        i += 1
        m += n
    return m

k=int(input('请输入相加整数个数'))
my_mul(k)

        
        
        
```
