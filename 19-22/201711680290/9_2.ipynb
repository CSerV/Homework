### 2、随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。

```python
import random
A=[]
V=[]
for i in range(1,100001):
    A.append(i)
    
for i in range(10000):
    v = random.choice(A)
    V.append(v) 
d = {k:V.count(k) for k in set(V)}

print(d)
```

```python
def maa():
    with open(r'd:\temp\test.txt', 'w') as fw, open(r'd:\temp\语料.txt') as fr:
        line_number = 0
        for line in fr:
            if line_number < 5001:
                fw.write(line)
                line_number += 1
            else:
                break
    with open(r'd:\temp\test.txt', 'w') as f:
        text = f.read()
    words = [word.split('/')[0] for word in text.split()]
    with open(r'd:\temp\test1.txt', 'w') as fw1, open(r'd:\temp\语料.txt') as fr:
        line_number = 0
        for line in fr:
            if line_number > 5000 and line_number < 10001:
                fw1.write(line)
                line_number += 1
            else:
                break
    with open(r'd:\temp\test1.txt', 'w') as f1:
        text1 = f1.read()
    words1 = [word.split('/')[0] for word in text1.split()]
    A=[]
    B=[]
    for i in words1:
        if i in words:
            A.append(i)        
        else:
            B.append(i)
    print('共用字',A)
    print('独用字',B)
print(maa()) 
```

### 4、挑战性任务：统计本任务中的语料.txt文件，得到每一个词对应不同词性的频次，可表示为：{词:[['n',1000], ['v', 98]...],...}，即在dict对象中，key为词，value为列表，输出词频最高的50个条目。

```python
from collections import Counter

def count_words_freq_dict(filename):
    words_freq_dict = Counter()
    
    with open(filename) as f:
        for line in f:
            words_freq_dict.update([word.split('/')[0] for word in line.split()])
    return words_freq_dict

def main():
    filename = r'd:\temp\语料.txt'
    table = count_words_freq_dict(filename)
    print(table.most_common(50))

if __name__ == '__main__':
    main()
```
