```python
'''3、对本任务中的语料.txt文件，抽取其5001-10000行存为test1.txt文件
      写函数，可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）。'''
```

```python
import random

def st():
    n=random.randint(5001,10000)
    with open(r'c:\temp\语料.txt') as f:
        line=f.read()
    words=[word.split('/')[0] for word in line.split()]
    st1=random.sample(words,k=n)
    return st1
def final(outs):
    fh = open(r'c:\temp\test.txt', 'w')
    for out in outs:
        fh.write(out+"\n")
    fh.close
    from collections import Counter
    dict_outs=dict(Counter(outs))
    outs1=[]
    outs2=[]
    for item in dict_outs.items():
        if item[1]==1:
            outs1.append(item[0])
        else:
            outs2.append(item[0])
    print('''共用字有
{}


独用字有
{}'''.format(outs2,outs1))
outs=st()
final(outs)
```

```python
'''4、挑战性任务：统计本任务中的语料.txt文件，得到每一个词对应不同词性的频次
      可表示为：{词:[('n',1000), ('v', 98)...],...}，即在dict对象中，key为词，value为列表，输出词频最高的50个条目。'''
```

```python
with open(r'c:\temp\语料.txt') as f:
    line=f.read()
words=[word.split('/') for word in line.split()]#[['习近平', 'nrfg'], ['新年贺词', 'l'], ['为', 'p'], ['人民', 'n'], ['点', 'm']]
words2=[word2 for word2 in line.split()]#['习近平/nrfg', '新年贺词/l', '为/p', '人民/n', '点/m','！/x']
colkey=[]
for word in words:
    colkey.append(word[0])
from collections import Counter
listkeys=sorted(dict(Counter(colkey)).items(),key=lambda item:item[1])
keys_50=listkeys[:-51:-1]#[('，', 27285), ('的', 15590), ('。', 11646), ('、', 5262),..., ('；', 622), ('2014', 611), ('我们', 608)]
final=[]
for exkey in keys_50:
    key=list(exkey)
    final.append(key[0])#先把key放进去
    final.append([])#造一个放词性的列表
    for word2 in words2:
        if key[0] in word2[:len(key[0])]:#在word2里找key[0]，找到之后记住词性，加到final的跟在key[0]后面的一个记词性的列表里
            temp1=word2[len(word2)-len(key[0]):len(word2)]#记住词性，比如'习近平/nrfg'，temp1就是nrfg
            n=final.index(key[0])#n保存key[0]这个词在final列表的位置
            if temp1 in final[n+1]:#如果词性已经在列表里就加一
                mark=final[n+1].index(temp1)#找出那个词的位置
                final[n+1][mark+1]+=1
            elif temp1 not in final[n+1]:#词性没在词性列表就往后加
                final[n+1].append(temp1)
                final[n+1].append(1)

list1=[]
for i in range(50):
    temp1=(final[2*i],final[2*i+1])
    list1.append(temp1)
final2=dict(list1)
print(final2)
```
