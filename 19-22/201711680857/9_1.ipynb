```python
import random,time,bisect
A=[]
B=[]
C=[]
D=[]
E=[]
for i in range(1,100001):
    A.append(i)
B=A
random.shuffle(B)
for j in range(1000):
    a=random.randint(1,110000)
    C.append(a)

start_time1=time.time()
for k in B:
    if k in C:
        D.append(k)
end_time1=time.time()
B_time=end_time1-start_time1
print('在B中查找到的整数有：',D)
print('用顺序查找法在B中查找所用时间为：',B_time)
print('\n')
start_time2=time.time()
for k in C:
    if k <=100000:
        k=bisect.bisect_left(A,k)
        E.append(k)
end_time2=time.time()
A_time=end_time2-start_time2
print('在A中查找到的整数有：',E)
print('用二分查找法在A中查找所用时间为：',A_time)
```
