```python
import random,time,bisect
listA=[]
listB=[]
for i in range(1,100001):
    listA.append(i)
print(listA)
for i in range(1,100001):
    listB.append(i)
random.shuffle(listB)
print(listB)
```

```python
n=0
listC=[]
while n<1000:
    m=random.randint(1,110000)
    listC.append(m)
    n+=1
print(listC)
```

```python
listD=[]
start_time1=time.time()
for p in listC:
    if p in listB:
        listD.append(p)
end_time1=time.time()
print(listD)

listE=[]
start_time2=time.time()
for q in listC:
    a=bisect.bisect_left(listA,q)
    b=bisect.bisect_right(listA,q)
    if a!=0 and b!=0:
        listE.append(q)
end_time2=time.time()
print(listE)
```

```python
print(end_time1-start_time1,end_time2-start_time2)
```
