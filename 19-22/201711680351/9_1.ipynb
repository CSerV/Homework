```python
a=[]
for i in range(100000+1):
    for k in range(i):
        a.append(i)
        break
b = a[100000-1::-1]

import random, bisect
       
def loc(x):
    if x not in b:
        return -1
    else:
        for i in range(100000-1):
            while x == b[i]:
                print(x,'这个数在b中所处的位置是',i)
                break
    
def bi_search(m):
    print(m,'这个数在a中的位置是',bisect.bisect_left(a, m))
    
for i in range(1000):
    number = random.randint(1,110001)
    
    loc(number)
    bi_search(number)
```
