```python
#随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
```

```python
import random
q=[]
for i in range(10000):
    q.append(random.randint(0,100000))

dict={}
for o in q:
    if o not in dict:
        dict[o]=1
    else:
        dict[o]+=1

for key in sorted(dict.keys()):
    print(key, dict[key])
```
