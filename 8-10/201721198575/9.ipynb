# 2、随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。


```python
import random

numbers = {}
for i in range(0,10000):
    number = random.randint(0,100000)
    if number in numbers: 
        numbers[number] += 1
    else: 
        numbers[number] = 1

list=sorted(numbers,reverse = True)
for ele in list:
    print(ele,':',numbers[ele])
```

# 3、对本任务中的语料.txt文件，随机抽取其5001-10000行存为test1.txt文件，写函数，可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）。

```python
from collections import Counter

def random_file1(filename):
    i = 0
    fh = open(r'c:\1\text.txt','w')
    with open(filename) as f:
        for line in f:
            i = i + 1
            if i > 5000:
                fh.close()
                return r'c:\1\text.txt'
            fh.write(line)
            
def random_file2(filename):
    i = 0
    fh = open(r'c:\1\text1.txt','w')
    with open(filename) as f:
        for line in f:
            i = i + 1
            if i > 5000 and i < 10001:
                fh.write(line)
    fh.close()
    return r'c:\1\text1.txt'
    
def compare(filename1,filename2):
    dict1 = Counter()
    dict2 = Counter()
    same = {}
    differ = {}
    
    with open(filename1) as f1:
        for line in f1:
            dict1.update([word.split('/')[0] for word in line.split()])
    with open(filename2) as f2:
        for line in f2:
            dict2.update([word.split('/')[0] for word in line.split()])
    for word in dict2:
        if word in dict1:
            same[word] = 1
        else:
            differ[word] = 1
#     for word in dict2:
#         if word not in same:
#             if word not in differ:
#                 differ[word] = 1
                
    return same,differ

def main():
    filename = r'c:\语料.txt'
    filename1 = random_file1(filename)
    filename2 = random_file2(filename)
    # print(filename1,filename2)
    same,differ = compare(filename1,filename2)
    print('这两个文件的共用字为：')
    for key in same:            
        print(key,end='  ')
    print('\n-----------------------------------------------------------------------------------------------------------------------')
    print('这两个文件的独用字为：')
    for key in differ:            
        print(key,end='  ')
   
if __name__ == '__main__':
    main()
```
